# U.S. Medical Insurance Costs

```python
# will need for csv function like csv.DictReader
import csv
```

```python
# emtpy lists for various attributes
ages = []
sexs = []
bmis = []
num_children = []
smoker_status = []
regions = []
insurance_charges = []
```

```python
# function that will get all the data from the columns of the csv file into the designated list
def get_csv_data(list, column, csv_file):
    with open(csv_file, newline = "") as file:
        # convert the file into a dictionary object to handle the data
        convDict = csv.DictReader(file)
        # loops through the data in each row
        for row in convDict:
            list.append(row[column])
    return list
```

```python
get_csv_data(ages, 'age', 'insurance.csv')
get_csv_data(sexs, 'sex', 'insurance.csv')
get_csv_data(bmis, 'bmi', 'insurance.csv')
get_csv_data(num_children, 'children', 'insurance.csv')
get_csv_data(smoker_status, 'smoker', 'insurance.csv')
get_csv_data(regions, 'region', 'insurance.csv')
get_csv_data(insurance_charges, 'charges', 'insurance.csv')
```

```python
# function that finds the average age of a patient
def avg_age(age_list):
    total = 0
    for age in age_list:
        total += int(age)
    return total / len(age_list)

average_age_of_patients = int(avg_age(ages)) # converted to a whole number because age doesn't make sense as a decimal
print(f"The averge age of a patient is {average_age_of_patients} years old")
```

```python
# funtion that finds number of males vs females
def male_vs_female(sex_list):
    male_count = 0
    female_count = 0
    for gender in sex_list:
        if gender == "male":
            male_count += 1
        else:
            female_count += 1
    return male_count, female_count

num_male, num_female = male_vs_female(sexs)
print(f"There are {num_male} males vs {num_female} females.")
```

```python
# function that calculates the yearly medical charges of the patients
def avg_yearly_charge(charge_list):
    total_charge = 0
    for charge in charge_list:
        total_charge += float(charge)
    return total_charge / len(charge_list)

avg_charge = avg_yearly_charge(insurance_charges)
print(f"The average yearly medical charge of all patients is ${avg_charge:.2f}")
```

```python
# finds all the unique location of all the patients
def geo_location(region_list):
    unique_region = []
    for region in region_list:
        if region not in unique_region:
            unique_region.append(region)
    return unique_region

uni_region = geo_location(regions)
print("Locations of our patient are: ")
for i in range(len(uni_region)):
    if i != len(uni_region):
        print(uni_region[i])
```

```python
# creates a dictionary that contains all patients information
def dict_of_patients(age, sex, bmi, children, smoker, region, charge):
    dict_patients = {}
    for i in range(len(age)):
        dict_patients[i] = {"Age": age[i], "Sex": sex[i], "BMI": bmi[i], "Children": children[i], "Smoker": smoker[i], "Region": region[i], "Charge": charge[i]}
    return dict_patients
 
print(dict_of_patients(ages, sexs, bmis, num_children, smoker_status, regions, insurance_charges))
```
